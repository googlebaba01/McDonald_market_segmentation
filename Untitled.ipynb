```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
df = pd.read_csv("mcdonalds.csv")  # Reading the data set
```

```python
df.head()
```

```python
df.shape
```

```python
# Custom function to convert the binary variables into numerical values in order to perform analysis on the dataset.

def Convert_to_data(x):
    df[x] = LabelEncoder().fit_transform(df[x])
    return df

cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in cat:
    Convert_to_data(i)
df
```

```python
df.describe()
```

```python
df_first_eleven = df.loc[:,cat]
df_first_eleven
```

```python
x = df.loc[:,cat].values
x
```

```python
#This code block performs prinicpal Component analysis on the data and converting it into a custom data set
pca_data = preprocessing.scale(x)
pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
pf.describe()
```

```python
pca.explained_variance_ratio_ #This command tells the variance contained by the PCA components and we can see that only 3 PCA components contain the most information.
```

```python
# This code block creates a new dataset showing how much information is contained about each variable in each Principal Component.
loadings = pca.components_
num_pc = pca.n_features_in_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df_eleven.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
# This code block applies K-Means clustering over the dataset to find Market Segments
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df_first_eleven)
visualizer.show()
```

```python
# From the previous block of code we can find that the optimal value of number of clusters is 4, but I trained it will 8 clusters.
kmeans = KMeans(n_clusters=8, init='k-means++', random_state=0).fit(df_first_eleven)
df['cluster_num'] = kmeans.labels_ #adding to df
```

```python
df.head(10)
```

```python
df['Like']= df['Like'].replace({'I hate it!-5': '-5','I love it!+5':'+5'}) # Processing the values of the Like column
```

```python
df.head(10)
```

```python
# This is a heatmap showing the relationship between the Principal Components and the variables of the dataset.
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='gist_rainbow')
plt.show()
```

```python
# This block of code uses seaborns scatterplot to plot the data points and their centroids.
sns.scatterplot(data=pf, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="*", c="g", s=80, label="centroids")
plt.legend()
plt.show()
```

```python
# This block of code creates a crosstab dataset which shows the number of people in each market segment.
crosstab_gender =pd.crosstab(df['cluster_num'],df['Gender'])
crosstab_gender
crosstab_gender['Total']=(crosstab_gender['Female']+crosstab_gender['Male'])
```

```python
crosstab_gender
```

```python
# This block of code finds how many times people visit the respective market segments. We have named it as Visit Frequency.
df['VisitFrequency'] = LabelEncoder().fit_transform(df['VisitFrequency'])
visit = df.groupby('cluster_num')['VisitFrequency'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
df['Like'] = LabelEncoder().fit_transform(df['Like'])
Like = df.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
# This block of code tell us the ratio of male and female present in the market segments or clusters.
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
Gender = df.groupby('cluster_num')['Gender'].mean()
Gender = Gender.to_frame().reset_index()
Gender
```

```python
segment = Gender.merge(Like, on='cluster_num', how='left').merge(visit, on='cluster_num', how='left')
segment
```

```python
# In this block of code we have show the variation of  of different segment with respect to their visiting Frequency to McDonald's
plt.figure(figsize = (9,4))
sns.scatterplot(x = "VisitFrequency", y = "Like",data=segment,s=400, color="b")
plt.title("All Clusters Comparison in terms of Frequency of Visit",
          fontsize = 15) 
plt.xlabel("Visit", fontsize = 12) 
plt.ylabel("Like", fontsize = 12) 
plt.show()
```
